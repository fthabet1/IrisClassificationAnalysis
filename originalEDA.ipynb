```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('IRIS.csv')

print(df.head())

print("\nDataset Description")
print(df.describe())

print("\nDataset Null Values")
print(df.isnull().sum())
```

```python
plt.pie(df['species'].value_counts(), labels=df['species'].unique(), autopct='%1.1f%%')
plt.title('Species Distribution')
plt.show()
```

```python
setosa = df[df['species'] == 'Iris-setosa']
versicolor = df[df['species'] == 'Iris-versicolor']
virginica = df[df['species'] == 'Iris-virginica']

numberedSpecies = df.copy()

numberedSpecies['species'] = numberedSpecies['species'].map({'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2})
```

```python
sns.pairplot(data = numberedSpecies, hue='species', hue_order=[0, 1, 2], palette='Set1')
plt.show()
```

```python
sns.violinplot(data = df, x = df['petal_length'], hue = df['species'])
plt.title('Petal Length Distribution')
plt.show()

sns.violinplot(data = df, x = df['petal_width'], hue = df['species'])
plt.title('Petal Width Distribution')
plt.show()
```

### Observation: Iris-setosa petals are always the lowest in terms of length and width
